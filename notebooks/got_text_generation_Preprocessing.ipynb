### Preprocessing

```python
#import libraries
import os
from cleantext import clean
```

```python
file = open('../datasets/got1.txt', 'r')
# read all text
txtDataset = file.read()
# close the file
file.close()
```

```python
print(txtDataset[:200])
```

```python
#lets clean text using clean library 
txtDataset=clean(txtDataset,lower=True,normalize_whitespace=True,no_punct=True,no_digits=True, no_line_breaks=True)
print(txtDataset[:200])
```

```python
#now I'll convert the text into a tokenized list
lstTokens = txtDataset.split()
lstTokens
```

```python
print(f'Total Tokens: {len(lstTokens)}')
print(f'Unique Tokens: {len(set(lstTokens))}')
```

### Sequences
Let's organize the long list of tokens into sequences of 50 input words and 1 output word.
That is, sequences of 51 words. We can do this by iterating over the list of tokens from token 51
onwards and taking the prior 50 tokens as a sequence, then repeating this process to the end of
the list of tokens. We will transform the tokens into space-separated strings for later storage
in a file.

```python
length = 50 + 1
sequences = list()
for i in range(length, len(lstTokens)):
    # select sequence of tokens
    lstSeq = lstTokens[i-length:i]
    # convert into a line
    strLine = ' '.join(lstSeq)
    # store
    sequences.append(strLine)
print(f'Total Sequences: {len(sequences)}')
```

### Save the sequences
Save the sequences into a file, in order to use in the modelling stage

```python
txtSequence = '\n'.join(sequences)
file = open('../datasets/got1_sequences.txt', 'w')
file.write(txtSequence)
file.close()
```
